```python
# Import dependencies
import pandas as pd
```

```python
# Read 2022 Jun data from csv to dataframe
file_path_jun = "data/JC-202206-citibike-tripdata.csv"
data_2022_jun = pd.read_csv(file_path_jun)

# Read 2022 Jul data from csv to dataframe
file_path_jul = "data/JC-202207-citbike-tripdata.csv"
data_2022_jul = pd.read_csv(file_path_jul)
```

```python
data_2022_jun.head()
```

```python
data_2022_jul.head()
```

```python
# Concatenate Jun and Jul data - 2022
data_2022_jun_jul = pd.concat([data_2022_jun,data_2022_jul],axis=0)
print(data_2022_jun.ride_id.count())
print(data_2022_jul.ride_id.count())
print(data_2022_jun_jul.ride_id.count())
```

```python
# Get column names
data_2022_jun_jul.columns
```

```python
# Write the processed data to a csv file
file_path_2022 = "data/nyc_citibike_data_2022_jun_jul.csv"
data_2022_jun_jul.to_csv(file_path_2022)
```
